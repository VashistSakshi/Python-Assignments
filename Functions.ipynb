1. What is the difference between a function and a method in Python?

- A function is a block of reusable code that performs a specific task. It is defined independently using the def keyword and can be called by its name anywhere in the program.

  A method is similar to a function but is associated with an object or a class. It is invoked using dot notation and usually works with the data stored in that object.

  **Key difference:** Functions are independent, while methods are tied to objects.

2. Explain the concept of function arguments and parameters in Python.

- Parameters are the variables that are listed in a function definition. They act as placeholders for values.

    Arguments are the actual values supplied to the function when it is called.

    **Example** in words: If a function is defined as def greet(name):, then name is the parameter. When we call greet("Alice"), the string "Alice" is the argument.

3. What are the different ways to define and call a function in Python?

- Normal functions: Defined using the def keyword and called by their name.

  Lambda (anonymous) functions: Small, one-line functions defined using the lambda keyword, often used for simple tasks.

  Built-in functions: Functions provided by Python, such as len(), sum(), print().
  Thus, functions can be user-defined or already available in Python.

4. What is the purpose of the return statement in a Python function?

- The return statement is used to send a result back from the function to the place where it was called.

  Without return, a function will perform its task but give back nothing (or None).

  Using return makes functions reusable because the result can be stored in variables, used in expressions, or passed to other functions.

5. What are iterators in Python and how do they differ from iterables?

- An iterable is any object that can return its elements one by one, such as lists, tuples, or strings.

  An iterator is an object created from an iterable that produces elements sequentially using the next() method until all elements are exhausted.

  Difference: Iterables can be looped over multiple times, but iterators remember their current position and can only be consumed once unless recreated.

6. Explain the concept of generators in Python and how they are defined.

- Generators are a special type of function that produce a sequence of values lazily using the yield keyword.

  Unlike normal functions, which return a single value and end, generators can pause their execution and resume later, producing one value at a time.

  They are useful when dealing with large amounts of data because they donâ€™t store everything in memory at once.

7. What are the advantages of using generators over regular functions?

- 1. Memory efficiency: Generators generate values one at a time, avoiding the need to store large sequences in memory.

  2. Lazy evaluation: Values are produced only when requested, which saves resources.

  3. Convenience: They are easier to write than creating custom iterator classes.

  4. Performance: Generators are generally faster for large data processing compared to creating full lists.

**8**. What is a lambda function in Python and when is it typically used?

- A lambda function is a small, anonymous function created using the lambda keyword.

  It can take any number of arguments but can only contain a single expression.

  It is typically used for short tasks such as sorting, filtering, or mapping, where defining a full function would be unnecessary.

9. Explain the purpose and usage of the map() function in Python.

- The map() function applies a given function to each element of an iterable and returns a new iterable (map object) with the results.

  Its main purpose is data transformation, for example: converting a list of strings to integers, or squaring each number in a list.

10. What is the difference between map(), reduce(), and filter() functions in Python?

- map(): Applies a function to each element of an iterable and returns the transformed results.

  filter(): Applies a condition to each element and returns only those elements that satisfy the condition (True).

  reduce(): Applies a function cumulatively to the elements of an iterable, reducing them to a single result (like calculating the total sum or product).

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

for num in range(1, 201):
    if is_prime(num):
        print(num)
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
def fibonacci(n):
    a, b = 0, 1  # first two Fibonacci numbers
    count = 0
    while count < n:
        yield a    # produce the next Fibonacci number
        a, b = b, a + b  # update for next number
        count += 1

# Example usage:
for num in fibonacci(10):
    print(num, end=" ")
```

 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(n):
    for i in range(n+1):
        yield 2**i

print(list(powers_of_two(5)))
```

 7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_lines(filename):
    with open(filename, "r") as file:
        for line in file:
            yield line.strip()

for line in read_file_lines("filename along with path"):
    print(line)
```

 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

```python
data = [(1, 5), (2, 2), (3, 9), (4, 1)]
sorted_data = sorted(data, key=lambda x: x[1])
print(sorted_data)
```

 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
celsius = [0, 10, 20, 30, 40]
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))
print(fahrenheit)
```

 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def remove_vowels(s):
    return "".join(filter(lambda x: x.lower() not in "aeiou", s))

print(remove_vowels("hello world"))
```

 11. Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,- â‚¬ if the value of the
order is smaller than 100,00 â‚¬.
 Write a Python program using lambda and map.

```python
orders = [
    [34587, 4.25, 100],
    [98762, 5.0, 50],
    [77226, 10.0, 5],
    [88112, 2.0, 200]
]

result = list(map(lambda x: (x[0], x[1]*x[2] if x[1]*x[2] >= 100 else x[1]*x[2] + 10), orders))

print(result)
```
